# Football Data Analysis

```python
# importing libraries for data analysis and visualization
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
DF = pd.read_excel("EPL_20_21 (1).xlsx")
```

```python
DF.head()
```

```python
DF.tail()
```

```python
DF.describe()
```

```python
DF.head()
```

```python
# Looking for the missing values in dataset
DF.isna().sum()
```

```python
# Creating 2 new columns 
DF["MinsPerMatch"] = (DF["Mins"] / DF["Matches"]).astype(int)
DF["GoalsPerMatch"] = (DF["Goals"] / DF["Matches"]).astype(float)
DF
```

```python
# Calculating total goals 

TotalGoals = DF["Goals"].sum()
TotalGoals
```

```python
# Calculating Penalty goals

PenaltyScored = DF["Penalty_Goals"].sum()
PenaltyScored
```

```python
# Penalty attempts

PenaltyAttempted = DF["Penalty_Attempted"].sum()
PenaltyAttempted
```

```python
# plottting a piechart for the penalties scored and penalties missed

plt.figure(figsize=(13,6))
PenaltyMissed = DF["Penalty_Attempted"].sum() - PenaltyScored
Data = [PenaltyMissed , PenaltyScored]
Labels = ["Penalties_Missed" , "Penalties_Scored"] 
Color = sns.color_palette("Set2")
plt.pie(Data, labels = Labels, colors = Color , autopct = "%.0f%%")
plt.show()
```

```python
# Unique Positions
DF["Position"].unique()
```

```python
# Total FW Players
FWPlayers = DF[DF["Position"] == "FW"]
```

```python
FWPlayers
```

```python
# Players From Different Nationalities
np.size((DF["Nationality"].unique()))
```

```python
# Most Players from a nation
Nationalities = DF.groupby("Nationality").size().sort_values(ascending = False)
Nationalities.head(10).plot(kind = "bar" , figsize=(12, 6) , color= sns.color_palette("rocket_r"))
```

```python
# Clubs with maximum number of players in their squad

DF["Club"].value_counts().nlargest(10).plot(kind="barh", color= sns.color_palette("magma"), figsize=(13, 6))
```

```python
# Clubs with least players in their squad

DF["Club"].value_counts().nsmallest(10).plot(kind="barh", color= sns.color_palette("magma"), figsize=(13, 6))
```

```python
# Players based on age group

under20 = DF[DF["Age"] <= 20]
aged20_25 = DF[(DF["Age"] > 20) & (DF["Age"] <=25)]
aged25_30 = DF[(DF["Age"] > 25) & (DF["Age"] <=30)]
above30 = DF[DF["Age"] >30]
```

```python
# Creating pie chart for the above data
x = np.array([under20["Name"].count(), aged20_25["Name"].count(), aged25_30["Name"].count(), above30["Name"].count()])
mylabels = ["<=20" , " > 20 & <=25" , " > 25 & <=30" , " > 30"]
plt.title("Total Players With Age" , fontsize=20)
plt.pie(x, labels = mylabels, autopct="%.1f%%")
plt.show()
```

```python
# All Players under 20 years of age

Players20 = DF[DF["Age"] < 20]
Players20["Club"].value_counts().plot(kind="barh" , color=sns.color_palette("cubehelix"), figsize=(12,6))
```

```python
# Average age of players in each club

plt.figure(figsize=(12, 6))
sns.boxplot(x="Club", y="Age", data=DF)
plt.xticks(rotation=90)
```

```python
NumOfPlayer = DF.groupby("Club").size()
data1 = (DF.groupby("Club")["Age"].sum())/ NumOfPlayer
data1.sort_values(ascending=False)
```

```python
DF.head()
```

```python
# Total assists from each club

AssistByClubs = pd.DataFrame(DF.groupby("Club", as_index=False)["Assists"].sum())
sns.set_theme(style="whitegrid", color_codes=True)
ax = sns.barplot(x="Club", y="Assists" , data=AssistByClubs.sort_values(by="Assists"),palette="Set2")
ax.set_xlabel("Clubs", fontsize=30)
ax.set_ylabel("Assists" , fontsize=20)
plt.xticks(rotation=75)
plt.rcParams["figure.figsize"] = (25,12)
plt.title("Plot of Clubs vs Total Assists", fontsize=20)
```

```python
# Top 10 Assists

Top10Assists = DF[["Name", "Club" , "Assists", "Matches"]].nlargest(n=10, columns = "Assists")
Top10Assists
```

```python
# Plottting bar chart for total goals vs Clubs

GoalsByClubs = pd.DataFrame(DF.groupby("Club", as_index=False)["Goals"].sum())
sns.set_theme(style="whitegrid", color_codes=True)
ax = sns.barplot(x="Club", y="Goals" , data=GoalsByClubs.sort_values(by="Goals"),palette="Set2")
ax.set_xlabel("Clubs", fontsize=30 , weight="bold")
ax.set_ylabel("Goals" , fontsize=20)
plt.xticks(rotation=75 , weight="bold" , fontsize=20)
plt.rcParams["figure.figsize"] = (25,12)
plt.title("Plot of Clubs vs Total Goals", fontsize=30)
```

```python
# Most goals by Players 

Top10Goals = DF[["Name" , "Club" , "Goals", "Matches"]].nlargest(n=10 , columns="Goals")
Top10Goals
```

```python
# Goals per match

Top10GoalsPerMatch = DF[["Name" , "GoalsPerMatch" , "Club" , "Matches" , "Goals"]].nlargest(n=10 , columns="Goals")
Top10GoalsPerMatch
```

```python
# Pie Chart for - Goals With Assists vs Goals Without Assists

plt.figure(figsize=(13, 6))
Assists = DF["Assists"].sum()
DATA = [TotalGoals - Assists, Assists]
labels1 = ["Goals With Assists" , "Goals Without Assists"]
color2 = sns.color_palette("Set1")
plt.pie(DATA , labels=labels1 , colors=color2 , autopct="%.0f%%")
plt.show()
```

```python
# Top 10 Players with most yellow cards 

DFYellow = DF.sort_values(by="Yellow_Cards", ascending=False)[:10]
plt.figure(figsize=(20, 6))
plt.title("Players With Most Yellow Cards")
c = sns.barplot(x=DFYellow["Name"], y=DFYellow["Yellow_Cards"] , label="Players" , color="yellow")
plt.ylabel("Number Of Yellow Cards")
c.set_xticklabels(c.get_xticklabels(), rotation=90)
c
```
